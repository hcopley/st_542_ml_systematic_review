```python
# Setup: In Google Colab use

# Import libraries
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
```

```python
# Load data
df = pd.read_csv('/test_data.csv')  # Your cleaned data from R
# Use the text column directly
texts = df['text']

```

```python
#create embeddings
model = SentenceTransformer('NeuML/pubmedbert-base-embeddings')
embeddings = model.encode(texts, batch_size=64, show_progress_bar=True)
```

```python
#Save as CSV
embeddings_df = pd.DataFrame(embeddings)
embeddings_df.to_csv('pubmedbert_embeddings.csv', index=False)
df.to_csv('data_for_comparison.csv', index=False)

print(f"Embeddings shape: {embeddings.shape}")
```
